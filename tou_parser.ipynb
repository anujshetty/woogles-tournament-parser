```python
from selenium import webdriver
import pandas as pd
from collections import defaultdict
from tabulate import tabulate

DRIVER_PATH = '../chromedriver_win32/chromedriver'
driver = webdriver.Chrome(executable_path=DRIVER_PATH)
driver.get('https://woogles.io/tournament/anujtest3')
```

```python
players=players_B
mapping=mapping_B
# Click left till Round 1
while True:
    try:
        driver.find_element_by_css_selector("[aria-label=left]").click()
    except Exception:
        break
games = {}
for player in players:
    games[player] = []

while True:
    rows = driver.find_elements_by_class_name('ant-table-row')
    for row in rows:
        text_arr = row.text.split('\n')
        split = text_arr[0].lower().split(' ')
        p1 = split[0]
        if len(split) > 1:
            status = split[1]
            if status.find('bye') >=0:
                games[p1].extend(['2001', players.index(p1)+1])
                continue
            if status.find('forfeit') >=0:
                games[p1].extend(['1000', players.index(p1)+1])
                continue
        _, p2, _, _, score1, score2 = text_arr[:6]
        p2 = p2.split(' ')[0].lower()

        win1, win2 = ('2','') if score1 > score2 else (('1', '1') if score1 == score2 else ('', '2'))

        games[p1].extend([win1 + str(score1), players.index(p2)+1])
        games[p2].extend([win2 + str(score2), players.index(p1)+1])

    try:
        driver.find_element_by_css_selector("[aria-label=right]").click()
    except Exception:
        break
```

```python
tabulate.PRESERVE_WHITESPACE = True
df = pd.DataFrame(games).T.reset_index()
df['index'] = df['index'].apply(lambda x: mapping[x].ljust(24, ' ')) #tsh allots about 21 characters to the name
df[0] = df[0].apply(lambda x: x.rjust(6, ' '))
#str
df
```

```python
div = 'B'
content = tabulate(df.values.tolist(), list(df.columns), tablefmt="plain")
start = f'*M01.22.2022 KSSA Virtual Jan Division {div}\n' \
        f'*{div}\n' \
        f'                                      0\n'
end = f'\n*** END OF FILE ***'
content = start + content[content.find('\n')+1:] + end
open('c:\Aupair\\np.tou', "w").write(content)
```
